### Informações básicas

Dado o problema de negócio que estamos avaliando, podemos considerar que o erro mais grave é prever que alguém é "bom pagador" quando na verdade é "mau pagador".

```python
from ucimlrepo import fetch_ucirepo
import pandas as pd
```

```python
german = fetch_ucirepo(id=144)
```

```python
german.metadata
```

```python
df = pd.concat([german.data.features, german.data.targets], axis=1)
df.head()
```

```python
columns = [
    "status_account",
    "duration",
    "credit_history",
    "purpose",
    "credit_amount",
    "savings",
    "employment_duration",
    "installment_rate",
    "personal_status_sex",
    "other_debtors",
    "residence_duration",
    "property",
    "age",
    "other_installment",
    "housing",
    "existing_credits",
    "job",
    "people_liable",
    "telephone",
    "foreign_worker",
    "target"
]

df.columns = columns
```

```python
df.shape
```

```python
df.info()
```

```python
df.describe()
```

```python
df["target"].value_counts()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(x="target", data=df)

plt.title("Distribuição das Classes")
plt.show()
```
